Arti dari masing-masing korelasi : 

- 1 yakni menunjukkan bahwa korelasi positif sempurna
    - Ketika satu variabel meningkat, variabel lainnya juga meningkat secara proporsional. Misalnya, jika variabel A dan variabel B memiliki korelasi 1, maka setiap kali A bertambah 1 unit, B juga akan bertambah sejumlah yang sama.
- 0 yakni menunjukkan bawa tidak adanya korelasi
    - Perubahan pada satu variabel tidak berpengaruh pada variabel lainnya. Contoh: Jika variabel A dan B memiliki korelasi 0, maka tidak ada pola yang konsisten dalam cara mereka berubah satu sama lain.
- -1 yakni menunjukkan bahwa korelasi negatif sempurna
    - Ketika satu variabel meningkat, variabel lainnya menurun secara proporsional. Misalnya, jika variabel A memiliki korelasi -1 dengan variabel B, maka setiap kali A bertambah 1 unit, B akan berkurang sejumlah yang sama.

### **QUANTITATIVE TEST**

```python
# Contoh penggunaan Quantitative Test yakni pada pencarian FREKUENSI pada suatu kolom tertentu
# Saya mengambil frekuensi dari tiap Genre yang muncul pada DataFrame IMDB tersebut

imdb_genre_freq = imdb_clear['Genre'].value_counts().reset_index()
imdb_genre_freq.columns = ['Genre', 'Frekuensi']
imdb_genre_freq
```

### **VISUALIZATION**

### **Visualisasi Frekuensi Kemunculan Genre Film pada DataFrame IMDB dengan Bar Chart**

```python
plt.figure(figsize=(12, 12), facecolor='skyblue')
plt.barh(imdb_genre_freq['Genre'], imdb_genre_freq['Frekuensi'], color='lightpink')
plt.xlabel('Frekuensi Kemunculan')
plt.ylabel('Genre Film')
plt.title('Frekuensi Genre Film pada DataFrame IMDB')
plt.gca().invert_yaxis()  
plt.show()

print("Dapat disimpulkan bahwa Film dengan Genre Animation, Adventure, Comedy merupakan Genre Film dengan Frekuensi terbanyak yang muncul pada DataFrame IMDB sebanyak 6 kali.")
```

### **Viasualisasi Film dengan IMDB Ratings tertinggi pada DataFrame IMBD**

```python
plt.figure(figsize=(12, 16))
plt.barh(imdb_clear['Title'], imdb_clear['imdbRating'], color='skyblue')
plt.xlabel('IMDb Rating')
plt.title('IMDb Ratings dari Film pada DataFrame IMDB : ')
plt.xlim(7.0, 10.0)
plt.gca().invert_yaxis()  
plt.show()

print("Dapat disimpulkan bahwa Film dengan IMDB Ratings tertinggi pada DataFrame IMDB yakni Film dengan judul 'The Dark Knight', dengan sebesar 9.0")
```

### **Visualisasi dari Sutradata dengan Runtime tertinggi dan terendah pada DataFrame IMDB**

```python
# Mengubah tipe data dari kolom 'Runtime' yang semula string menjadi integer dan menghilangakn objek ' min'

imdb_clear['Runtime'] = imdb_clear['Runtime'].str.replace(' min', '').astype(int)
```

```python
# Mencari sutradara dengan runtime Film tertinggi dan terendah pada DataFrame IMDB

runtime_sut = imdb_clear.groupby('Director')['Runtime'].mean().reset_index()

# Mendapatkan sutradara dengan runtime tertinggi dan terendah pada DataFrame IMDB
runtime_sut_max = runtime_sut.loc[runtime_sut['Runtime'].idxmax()]
runtime_sut_min = runtime_sut.loc[runtime_sut['Runtime'].idxmin()]
```

```python
sutradara = [runtime_sut_max['Director'], runtime_sut_min['Director']]
runtimes_data = [runtime_sut_max['Runtime'], runtime_sut_min['Runtime']]

plt.figure(figsize=(8, 6))
plt.barh(sutradara, runtimes_data, color=['green', 'red'])
plt.xlabel('Rata-Rata Runtime (Menit)')
plt.title('Sutradara dengan Rata-Rata Runtime Tertinggi dan Terendah pada DataFrame IMDB')
plt.show()
```

# **ANALISIS STATISTIK**

```python
# Count Mean (Kolom Numerik) // y.mean()

mean_num = {} # Menampung Kolom Numerik

for kolom in imdb_data.columns : # Looping untuk melakukan Count Mean pada Setiap Kolom Numerik pada DataSet
    if pd.api.types.is_numeric_dtype(imdb_data[kolom]) : # Memeriksa apakah kolom adalah tipe numerik
        mean_num[kolom] = imdb_data[kolom].mean() # Menghitung rata-rata dan menyimpannya dalam dictionary

# Membuat DataFrame
mean_num_df = pd.DataFrame(list(mean_num.items()), columns = ["Kolom", "Mean"])
mean_num_df
```

```python
# Count Variance (Kolom Numerik) // y.var(ddof = 1)

var_num = {} # Menampung Kolom Numerik

for kolom in imdb_data.columns : # Looping untuk melakukan Count Var pada Setiap Kolom Numerik pada DataSet
    if pd.api.types.is_numeric_dtype(imdb_data[kolom]) :
        var_num[kolom] = imdb_data[kolom].var(ddof = 1) # Menghitung Variance dan menyimpannya dalam dictionary

var_num_df = pd.DataFrame(list(var_num.items()), columns = ["Kolom", "Variance"])
var_num_df
```

```python
# Count Standar Deviation (Kolom Numerik) // y.std(ddof = 1)

stdev_num = {} # Menampung Kolom Numerik

for kolom in imdb_data.columns : # Looping untuk melakukan Count St Dev pada Setiap Kolom Numerik pada DataSet
    if pd.api.types.is_numeric_dtype(imdb_data[kolom]) :
        stdev_num[kolom] = imdb_data[kolom].std(ddof = 1) # Menghitung St Dev dan menyimpannya dalam dictionary

stdev_num_df = pd.DataFrame(list(stdev_num.items()), columns = ["Kolom", "Standar Deviation"])
stdev_num_df
```

```python
# Count Range (Kolom Numerik) // y.maxx() - y.min()

range_num = {} # Menampung Kolom Numerik

for kolom in imdb_data.columns : # Looping untuk melakukan Count Range pada Setiap Kolom Numerik pada DataSet
    if pd.api.types.is_numeric_dtype(imdb_data[kolom]) :
        range_num[kolom] = imdb_data[kolom].max() - imdb_data[kolom].min() # Menghitung Range dan menyimpannya dalam dictionary

range_num_df = pd.DataFrame(list(range_num.items()), columns = ["Kolom", "Count Range"])
range_num_df
```

```python
# Five Number Summary

five_number_summary = {} # Menampung Kolom Numerik

for kolom in imdb_data.select_dtypes(include=[np.number]).columns : # Looping untuk melakukan Summary pada Setiap Kolom Numerik pada DataSet
    summary = np.percentile(imdb_data[kolom].dropna(), [0, 25, 50, 75, 100])
    five_number_summary[kolom] = summary # Menghitung Summary dan menyimpannya dalam dictionary

five_number_summary_df = pd.DataFrame(five_number_summary, index = ["Min", "Q1", "Median", "Q3", "Max"]).T
five_number_summary_df
```

```python
# Box Plot Setiap Kolom Numerik

imdbkolom_numerik = imdb_data.select_dtypes(include="number").columns # Mengambil nama kolom yang memiliki tipe data numerik

plt.figure(figsize=(20, 15)) # Menentukan ukuran untuk plot

for i, kolom in enumerate(imdbkolom_numerik) : # Looping untuk membuat subplot box plot untuk setiap kolom numerik
    plt.subplot(len(imdbkolom_numerik), 1, i + 1)  # Menentukan posisi subplot
    sns.boxplot(y = imdb_data[kolom]) # Membuat boxplot untuk kolom yang sedang diiterasi
    plt.title(f"Boxplot Kolom {kolom}")  

plt.tight_layout()  
plt.show()  
```

```python
# Displot/Hisplot Setiap Kolom Numerik

imdbkolom_numerik = imdb_data.select_dtypes(include="number").columns # Mengambil nama kolom yang memiliki tipe data numerik

plt.figure(figsize=(20, 15)) 

for i, kolom in enumerate(imdbkolom_numerik) : # Looping untuk membuat subplot histogram untuk setiap kolom numerik
    plt.subplot(len(imdbkolom_numerik), 1, i + 1)  
    sns.histplot(imdb_data[kolom], kde=False, bins=30)  
    plt.title(f"Displot Kolom {kolom}")  

plt.tight_layout()  
plt.show()  
```

```python
# Mencari Median pada Setiap Kolom Numerik

from statistics import median  # Import fungsi median dari modul statistics

med_num = imdb_data.select_dtypes(include = "number").columns # Mengambil nama kolom yang memiliki tipe data numerik

for kolom in med_num : # Looping untuk menghitung median dari setiap kolom numerik
    kolom_data = imdb_data[kolom].dropna()    # Mengambil data kolom dan menghapus nilai NaN pada kolom
    if len(kolom_data) > 0 :  # Memastikan kolom tidak kosong setelah menghapus NaN
        med = median(kolom_data)  # Hitung median dari data kolom
        print("Median of %s is %s" % (kolom, med)) 
```

```python
# Mencari Modus pada Setiap Kolom Numerik

from statistics import mode # Import fungsi modus dari modul statistics

mode_num = imdb_data.select_dtypes(include = "number").columns # Mengambil nama kolom yang memiliki tipe data numerik

for kolom in mode_num : # Looping untuk menghitung modus dari setiap kolom numerik
    kolom_data = imdb_data[kolom].dropna()  # Mengambil data kolom dan menghapus nilai NaN pada kolom
    if len(kolom_data) > 0 :  # Memastikan kolom tidak kosong setelah menghapus NaN
        mod = mode(kolom_data) # Hitung modus dari data kolom
        print("Mode of %s is %s" % (kolom, mod)) 
```

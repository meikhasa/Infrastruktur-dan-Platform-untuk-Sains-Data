## **26.) EKSPLORASI DATA**

```python
genangan_df = pd.read_csv("C:/Users/Mahisa Ayu S/Downloads/data genangan air Surabaya1.csv", sep=",")
genangan_df
```

```python
# Mengecek tipe data pada tiap kolom pada DataFrame

data_types = genangan_df.dtypes
print("Tipe data per kolom : ")
print("\n")
data_types
```

```python
# Mencari jumlah Missing Value pada setiap kolom pada DataFrame

print("DATA GENANGAN : ")
print("\n")
display(genangan_df.isnull().sum())
```

```python
# Mengganti nilai NaN dengan nilai )

genangan_df_new = genangan_df.fillna(0)
genangan_df_new
```

```python
# Mencari jumlah Missing Value pada setiap kolom pada DataFrame

print("DATA GENANGAN (BARU) : ")
print("\n")
display(genangan_df_new.isnull().sum())
```

### **a.) Nilai Maksimum, Minimum, Rata-Rata dari Luas Genanagan, Tinggi Genangan dan Lama Genangan Air**

```python
# Nilai Maksimum

maksimum = {}

for kolom in genangan_df_new.columns : 
    if pd.api.types.is_numeric_dtype(genangan_df_new[kolom]) : 
        maksimum[kolom] = genangan_df_new[kolom].max()

maks_df = pd.DataFrame(list(maksimum.items()), columns = ["Kolom", "Nilai Maksimum"])
maks_df
```

```python
# Niali Minimum

minimum = {}

for kolom in genangan_df_new.columns : 
    if pd.api.types.is_numeric_dtype(genangan_df_new[kolom]) : 
        minimum[kolom] = genangan_df_new[kolom].min() 

min_df = pd.DataFrame(list(minimum.items()), columns = ["Kolom", "Nilai Minimum"])
min_df
```

```python
# Niali Minimum

mean = {}

for kolom in genangan_df_new.columns : 
    if pd.api.types.is_numeric_dtype(genangan_df_new[kolom]) : 
        mean[kolom] = genangan_df_new[kolom].mean() 

mean_df = pd.DataFrame(list(mean.items()), columns = ["Kolom", "Nilai Mean"])
mean_df
```

### **b.) Pie Chart untuk Membandingkan Banyak Peristiwa Genangan Berdasarkan Rayonnya**

```python
rayon_counts = genangan_df['Rayon'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(rayon_counts, labels=rayon_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Perbandingan Peristiwa Genangan Berdasarkan Rayon')
plt.axis('equal')  
plt.show()
```

### **c.) Hubungan antara Luas Genangan dan Skor Genangan, dan Apakah ada Hubungan antara Luas Genangan Air dan Lama Genangan Air?**

```python
# Menghitung matriks korelasi antar kolom numerik

genangan_korelasi = genangan_df_new[['Luas genangan', 'Skor Lama Genangan']].corr()
print("Matriks Korelasi dari Luas Genangan dan Skor Lama Genangan :")
genangan_korelasi
```

```python
plt.scatter(genangan_df_new['Luas genangan'], genangan_df_new['Skor Lama Genangan'])
plt.title('Scatter Plot of Luas Genangan vs Skor Lama Genangan')
plt.xlabel('Luas Genangan')
plt.ylabel('Skor Lama Genangan')
plt.show()
```

```python
# Menghitung matriks korelasi antar kolom numerik

genangan_korelasi1 = genangan_df_new[['Luas genangan', 'Lama genangan']].corr()
print("Matriks Korelasi dari Luas Genangan dan Lama Genangan :")
genangan_korelasi1
```

```python
plt.scatter(genangan_df_new['Luas genangan'], genangan_df_new['Lama genangan'])
plt.title('Scatter Plot of Luas Genangan vs Lama Genangan')
plt.xlabel('Luas Genangan')
plt.ylabel('Lama Genangan')
plt.show()
```
